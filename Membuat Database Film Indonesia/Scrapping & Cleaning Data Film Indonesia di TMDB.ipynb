```python
for ind, m_id in enumerate(movie_id):
    print(ind, 'from', len(movie_id))
    try :
        # Data json detail film menggunakan movie_id
        m = movie.details(m_id)
        # movie id
        id_movie.append(m_id)
        # imdb id film
        try:
            imdb_id.append(m.imdb_id)
        except:
            imdb_id.append(None)
        # Judul film
        try:
            title.append(m.title)
        except:
            title.append(None)
        # Ringkasan film
        try:
            overview.append(m.overview)
        except:
            overview.append(None)
        # Skor popularitas film
        try:
            popularity.append(m.popularity)
        except:
            popularity.append(None)
        # Keterangan adult film
        try:
            adult.append(m.adult)
        except:
            adult.append(None)
        # Bahasa asli film
        try:
            original_language.append(m.original_language)
        except:
            original_language.append(None)
        # Durasi film dalam menit
        try:
            runtime.append(m.runtime)
        except:
            runtime.append(None)
        # Pendapatan film
        try:
            revenue.append(m.revenue)
        except:
            revenue.append(None)
        # Biaya pembuatan film
        try:
            budget.append(m.budget)
        except:
            budget.append(None)
        # Tanggal rilis film
        try:
            release_dates.append(m.release_dates)
        except:
            release_dates.append(None)
        try:
            release_date.append(m.release_date)
        except:
            release_date.append(None)
        # Jumlah vote di themoviedb.org
        try:
            vote_count.append(m.vote_count)
        except:
            vote_count.append(None)
        # Rerata vote di themoviedb.org
        try:
            vote_average.append(m.vote_average)
        except:
            vote_average.append(None)
        # Sertifikasi film
        try:
            certifications.append(m.release_dates.results[0].release_dates[0].certification)
        except:
            certifications.append(None)

        # Data genre film : Satu film bisa direferensikan sebagai lebih dari satu genre
        try: 
            genres.append(m.genres)
        except:
            genres.append(None)
        try:
            genres_ = movie.details(m_id).genres
            genre_name_list.append([x.name for x in genres_])
        except:
            genre_name_list.append(None)
        try:
            genres_ = movie.details(m_id).genres
            genre_id_list.append([x.id for x in genres_])
        except:
            genre_id_list.append(None)
        
        # Negara asal film
        try:
            countries.append(m.production_countries)
        except:
            countries.append(None)
        try:
            countries_ = movie.details(m_id).production_countries
            country_name_list.append([x.name for x in countries_])
        except:
            country_name_list.append(None)
        try:
            countries_ = movie.details(m_id).production_countries
            country_id_list.append([x.iso_3166_1 for x in countries_])
        except:
            country_id_list.append(None)
        
        # Perusahaan produsen film
        try:
            companies.append(m.production_companies)
        except:
            companies.append(None)
        try:
            companies_ = movie.details(m_id).production_companies
            company_name_list.append([x.name for x in companies_])
        except:
            company_name_list.append(None)
        try:
            companies_ = movie.details(m_id).production_companies
            company_id_list.append([x.id for x in companies_])
        except:
            company_id_list.append(None)
        try:
            companies_ = movie.details(m_id).production_companies
            company_oc_list.append([x.origin_country for x in companies_])
        except:
            company_oc_list.append(None)
        
        # Bahasan terjemahan fil
        try:
            translations.append(m.translations)
        except:
            translations.append(None)
        try:
            translations_ = movie.details(m_id).translations.translations
            translation_name_list.append([x.english_name for x in translations_])
        except:
            translation_name_list.append(None)
        try:
            translations_ = movie.details(m_id).translations.translations
            translation_id_list.append([x.iso_3166_1 for x in translations_])
        except:
            translation_id_list.append(None) 
            
        
        # Artis pemeran Film  
        try:
            casts.append(m.casts)
        except:
            casts.append(None)
            
        try:
            casts_ = movie.details(m_id).casts.cast
            cast_oname_list.append([x.original_name for x in casts_])
        except:
            cast_oname_list.append(None)
        try:
            casts_ = movie.details(m_id).casts.cast
            cast_name_list.append([x.name for x in casts_])
        except:
            cast_name_list.append(None)
        try:
            casts_ = movie.details(m_id).casts.cast
            cast_character_list.append([x.character for x in casts_])
        except:
            cast_character_list.append(None)
        try:
            casts_ = movie.details(m_id).casts.cast
            cast_id_list.append([x.id for x in casts_])
        except:
            cast_id_list.append(None)
        try:
            casts_ = movie.details(m_id).casts.cast
            cast_popularity_list.append([x.popularity for x in casts_])
        except:
            cast_popularity_list.append(None)
    except:
        continue
```

### Data Pemeran

```python
# Melihat kolom data untuk mengambil kolom-kolom berkaitan dengan data pemeran
df_raw.columns
```

```python
# Mengumpulkan data pemeran dalam list
id_pemeran = []
nama_pemeran = []
nama_asli_pemeran = []
karakter_diperankan = []
popularitas_pemeran = []

for i in df_raw.cast_id_list:
    id_pemeran = id_pemeran+i
for i in df_raw.cast_name_list:
    nama_pemeran = nama_pemeran+i
for i in df_raw.cast_original_name_list:
    nama_asli_pemeran = nama_asli_pemeran+i
for i in df_raw.cast_character_list:
    karakter_diperankan = karakter_diperankan+i
for i in df_raw.cast_popularity_list:
    popularitas_pemeran = popularitas_pemeran+i
```

```python
# Mebuat kolom baru jumlah pemeran
df_raw['jumlah_pemeran'] =  df_raw['cast_id_list'].apply(lambda x: len(x))
```

```python
# Membuat list kosong id_film
id_film = []
# Mengumpulkan list id_film sejuamlah jumlah pemeran pada masing-masing film
for i, jp in zip(df_raw.id, df_raw['jumlah_pemeran']):
    id_film= id_film + [i]*jp
```

```python
# Membuat dataframe df_pemeran
df_pemeran = pd.DataFrame({'id'          : id_pemeran,
                        'nama'        : nama_pemeran,
                        'nama_asli'   : nama_asli_pemeran,
                        'id_film'     : id_film,
                        'karakter'    : karakter_diperankan, 
                        'popularitas' :popularitas_pemeran}) 
# Drop duplikasi data pada kolom id dan id_film
df_pemeran = df_pemeran.drop_duplicates(subset=['id', 'id_film'])
# Menampilkan ata df_pemeran
print(df_pemeran)
```

```python
# Menyimpan df_pemeran dalam bentuk file csv tanpa menyimpan indeksnya
df_pemeran.to_csv('pemeran.csv', index=False)
```
